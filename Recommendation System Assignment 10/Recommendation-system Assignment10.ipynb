```python
b_11=df.drop_duplicates(["Book.Title"]).reset_index(drop=True)
```

```python
bx=b_11.pivot(index="User.ID",
           columns="Book.Title",
           values="Book.Rating")
```

```python
bx1=bx.fillna(0,axis=1)
```

```python
bx1
```

```python
#calculating the sapital distances uisng cosine 
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine
import scipy
```

```python
bx_simir=1-pairwise_distances(bx1.values , metric='cosine')
```

```python
bx_simir
```

```python
bx_simir_df=pd.DataFrame(bx_simir)
```

```python
#lets  check for most similar users
bx_simir_df.idxmax()[0:5]
```

```python
df[(df["Book.Rating"]==7) | (df["Book.Rating"]==10)]
```

```python
#lets give recommendation
def recommend(users_id):
    temp=list(bx_simir_df.sort_values([users_id],ascending=True).head(10).index)
    books_list=[]
    for i in temp:
        books_list=books_list+list(df[df["User.ID"]==i]["Book.Title"])
    return  set(books_list)-set(df[df["User.ID"]==users_id]["Book.Title"])
```

```python
recommend(4)
```
